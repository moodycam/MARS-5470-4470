```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.arange(10)
a
```

```python
plt.plot(a, 3*a +5)
plt.xlabel('a')
plt.title('my plot')
plt.ylabel('3a+5')
plt.show()
```

```python
b = np.zeros((5,6))
b
```

```python
b[0,0] = 10
b[0,1] = 13
b
```

```python
plt.pcolormesh(b)
plt.colorbar()
```

```python
b[1 , : ] = 5
b
```

```python
plt.pcolormesh(b, cmap = 'inferno')
plt.colorbar()
```

```python
# a new array that is part of the old one
c = b [0:2,:]
c
```

```python
plt.pcolormesh(c)
plt.colorbar()
```
