```python
import matplotlib.pyplot as plt
```

```python
data = xr.open_dataset('C:\\Users\srv_veralab\Downloads\CESM.003.SST.1980.nc')
lat = np.array(data.lat)
lon = np.array(data.lon)
data = np.array(data.SST)
```

```python
meandata = np.mean(data, axis=0)
```

```python
fig, ax = plt.subplots()
p = ax.pcolormesh(lon, lat, meandata, cmap="autumn_r",
                  vmin=-2, vmax=30)
cb = plt.colorbar(p, extend="both")

# Aesthetics
cb.set_label('Sea Surface Temperature [$^{o}$C]')
ax.set_title('CESM Sea Surface Temperature (1980)')
ax.set_ylabel('Latitude')
ax.set_xlabel('Longitude')
```

```python
winter = data[(0,1,11),:,:]
```

```python
data_winter = np.mean(data[(0,1,11),:,:], axis=0)
```

```python
data_spring = np.mean(data[(2,3,4),:,:], axis=0)
```

```python
data_summer = np.mean(data[(5,6,7),:,:], axis=0)
```

```python
data_fall = np.mean(data[(8,9,10),:,:], axis=0)
```

```python
fig, ax = plt.subplots(figsize=(8,8), nrows=2, ncols=2, sharex=True,
                       sharey=True)

p = ax[0,0].pcolormesh(lon, lat, data_winter)
ax[0,1].pcolormesh(lon, lat, data_spring)
ax[1,0].pcolormesh(lon, lat, data_summer)
ax[1,1].pcolormesh(lon, lat, data_fall)

ax[0,0].set_title('Winter', fontsize=18)
ax[0,1].set_title('Spring', fontsize=18)
ax[1,0].set_title('Summer', fontsize=18)
ax[1,1].set_title('Fall', fontsize=18)


plt.colorbar(p, ax = ax, extend = "both")
```
