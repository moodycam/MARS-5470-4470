# Copies and views
    A slicing operation creates a view on the original array, which is just a way of accessing array data. Thus the original array is not copied in memory. 
    
    When modifying the view, the original array is modified as well:


```python
a=np.arange(10)
a
```

```python
b=a[::2]
b
```

```python
#Create a list of numbrs from 0 to 9
a = np.arange(10)
print("a: " + str(a))

#Creates a new list called b from every other number in list a
b = a[::2]
print("b: " + str(b))

#This changes the first element in b to 12
np.may_share_memory(a, b)
b[0] = 12
print("b[0] changed to 12: " + str(b))


#Inadvertantly, changes to b[0] also make changes to  a[0]
a
print("a[0] is now 12 too: " +str(a))
```

**Note: Changing the first element in array b changed the first element in array a**

```python
a=np.arange(10)
a
```

```python
c=a[::2].copy()
c
```
