# Line Plots

```python
#Generate data
t=np.linspace(-np.pi, np.pi, 64, endpoint=True)
c=np.cos(t)
s=np.sin(t)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(t,s)
ax.plot(t,c)
```

```python
fig, ax = plt.subplots(figsize=(10,10), ncols=2) 
ax[0].plot(t,s)
ax[1].plot(t,c)
#Anything with = is optional
```

```python
fig, ax = plt.subplots(figsize=(3,3)) 
ax.plot(t,s, label= 'sin', linewidth=1,color='red', marker='o', 
        markersize=3)
ax.plot(t,c, label='cosine', linewidth=3, color='#A9A9A9')
plt.legend()

```

```python
fig, ax = plt.subplots(figsize=(3,3)) 
ax.plot(t,s)
ax.plot(t,c)

#Now we want to mess with the axis..
#ax.set_title('Sines and Cosines', fontsize=18)
#ax.set_xlabel('Time')
#ax.set_ylabel('Magnitude')
#ax.set_ylim(-2,2)
#ax.set_xlim(-10,10)

ax.set(title='Sines and Cosines', xlabel('Time'), ylabel('Magnitude'), ylim(-2,2), xlim(-10,10))
```
