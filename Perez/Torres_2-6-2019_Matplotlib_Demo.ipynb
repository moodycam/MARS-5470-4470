# Line Plots

```python
# generate data
t = np.linspace(-np.pi, np.pi, 64, endpoint = True)
s = np.sin(t)
c = np.cos(t)
```

```python
fig, ax = plt.subplots(figsize = (3,3))
ax.plot(t,s)
```

```python
fig, ax = plt.subplots(figsize=(6,3), ncols=2)
```

```python
ax[0].plot[t,s]
ax[1].plot[t,c]
```

```python
fig, ax = plt.subplots(figsize=(3,3))
ax.plot(t,c, label='cosine', linewidth=3, linestyle='--', color = 'black')
ax.plot(t,s, label='sine', linewidth = 5, linestyle='-.', color = 'Red')
plt.legend()
```

```python
fig, ax = plt.subplots(figsize=(3,3))
ax.plot(t,c)
ax.plot(t,s)

###### 

#ax.set_title('Sines and Cosines', fontsize = 18)
#ax.set_xlabel('Time')
#ax.set_ylabel('Magnitude')
#ax.set_ylim((-2,2))
#ax.set_xlim((-10,10))
ax.set(title='sines and cosines', xlabel='Time', ylabel='Magnitude')
```
