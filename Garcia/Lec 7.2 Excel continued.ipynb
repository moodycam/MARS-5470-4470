Ex.1 From the movies spreadsheet in the above tutorial, make a histogram of all the years movies were made, combining the data in  three sheets. Make two plots, one with bins (bars)  of 10 years width, and one with bins of 20 years width. Label your axes and change the color of the bars. Interpret this data, i.e. describe the change in the number of movies made over time.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import xarray as xr
import pandas as pd
```

```python
excel_file = "movies.xls"
movies = pd.read_excel(excel_file)
movies.head()
```

```python
xlsx = pd.ExcelFile(excel_file)
movies_sheets = []
for sheet in xlsx.sheet_names:
    movies_sheets.append(xlsx.parse(sheet))
movies = pd.concat(movies_sheets)
movies.tail() # to check if my concatination worked
```

```python
sorted_by_year = movies.sort_values(['Year'], ascending=True)
sorted_by_year.head()
```

```python
movies['Year'].plot(kind='hist', bins=10, cmap = 'plasma')
plt.xticks(np.arange(1900, 2021, 10))
plt.title('Movies made per Decade')
plt.ylabel('Number of Movies')
plt.xlabel('Year')
```

```python
movies['Year'].plot(kind='hist', bins=20, cmap='inferno')
plt.xticks(np.arange(1900, 2021, 20))
plt.title('Movies made per 20 Years')
plt.ylabel('Number of Movies')
plt.xlabel('Year')
```

As the years went by, the amount of movies that were being made increased exponentially.
