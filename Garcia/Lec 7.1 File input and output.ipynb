E1. Plot the height function $H = -100(lat-40)^2 -400{(lon+106)^2} + 8000$ over 38 to 42 N and 108 to 104 W. What mountain range is this a rough approximation to?

```python
lat = np.linspace(38,43,50)
lon = np.linspace(-108,-104,50)
```

```python
lat, lon = np.meshgrid(lat, lon)
```

```python
plt.plot(lon, lat)
```

```python
H = -100*((lat-40)**2) - 400*((lon+106)**2)+8000 # makes the function
```

```python
plt.pcolormesh(lon, lat, H, cmap='plasma')
plt.colorbar()
```

```python
plt.contour(lon, lat, H)
```

```python
plt.contourf(lon, lat, H)
plt.colorbar()
```
